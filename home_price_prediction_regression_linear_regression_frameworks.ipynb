# Can you build a multivariate linear regression model that can predict the product sales based on the advertising budget allocated to different channels. The features are TV Budget ($),Radio Budget ($),Newspaper Budget ($) and the output is Sales (units)


# step 1: import the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

# step 2: reading the data and splitting it to input and output


```python
dataset = pd.read_csv('https://raw.githubusercontent.com/jeff-ai-ml/ml-datasets/refs/heads/main/home.csv')

print(dataset)
inputx = dataset.iloc[:, :-1].values
outputy = dataset.iloc[:, 2].values

print(inputx)
print(outputy)

#inputx = dataset.iloc[:, :-1].values
#outputy = dataset.iloc[:, 1].values
```

```python
print(dataset)
```

# step 3: select one thirds of the data for testing and two thirds for training

```python
input_train, input_test, output_train, output_test = train_test_split(inputx, outputy, test_size = 1/4, random_state = 7)
```

# step 4: selecting the simple Linear Regression model

```python
model = LinearRegression()
print("\nThe parameters of the model are\n\n",model.get_params())
#print(model.set_params())
print("\nThe model we are using is ", model.fit(input_train, output_train))

```

# step 5: testing or model prediction using testinput

```python
print("Hello! I can predict housing price for you")
square_feet = int(input("Enter the total square feet of house : "))
no_of_bedrooms = int(input("Enter the number of Bedrooms in the house : "))
testinput = [[square_feet, no_of_bedrooms]]
predicted_output = model.predict(testinput)
rounded_predicted_output = np.round(predicted_output)
print('\nThe prince of the house will be ',int(rounded_predicted_output[0]))
#yes = input("\nCan I proceed\n")
```

# step 6: Printing the testing results

```python
print("\nThe test input house-square-feet and number of bedrooms are as follows \n")
print(input_test)
# model predicting the Test set results
predicted_output = model.predict(input_test)
print("\nThe output (Price of House) \n")
print(predicted_output)
```

# step 7: Visualising the training results

```python
# Generate grid for prediction surface
square_feet_range = np.linspace(inputx[:, 0].min(), inputx[:, 0].max(), 20)
bedroom_range = np.linspace(inputx[:, 1].min(), inputx[:, 1].max(), 20)
sqft_grid, bed_grid = np.meshgrid(square_feet_range, bedroom_range)
grid_input = np.c_[sqft_grid.ravel(), bed_grid.ravel()]
price_grid = model.predict(grid_input).reshape(sqft_grid.shape)

# Create 3D plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Plot the prediction surface with color map
norm_prices = (price_grid - price_grid.min()) / (price_grid.max() - price_grid.min())
surf = ax.plot_surface(price_grid, bed_grid, sqft_grid,
                       facecolors=plt.cm.viridis(norm_prices),
                       rstride=1, cstride=1, alpha=0.9, antialiased=True)

# Plot actual data points
actual_prices = outputy
actual_bedrooms = inputx[:, 1]
actual_sqft = inputx[:, 0]
ax.scatter(actual_prices, actual_bedrooms, actual_sqft, color='black', label='Actual Data', s=50)

# Labels and title
ax.set_xlabel('Price ($)')
ax.set_ylabel('Number of Bedrooms')
ax.set_zlabel('Square Feet')
ax.set_title('3D House Price Prediction with Color Scale')

# Color bar for the surface
mappable = plt.cm.ScalarMappable(cmap=plt.cm.viridis)
mappable.set_array(price_grid)
cbar = plt.colorbar(mappable, ax=ax, shrink=0.5, aspect=10, pad=0.1)
cbar.set_label('Predicted Price ($)')

plt.legend()
plt.tight_layout()
plt.show()
```
